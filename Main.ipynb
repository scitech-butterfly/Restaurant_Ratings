```python
pr_cuisine = input("Enter your preference of cuisine: ")
pr_rating = float(input("Enter the average cost rating for two: "))  # Assuming the rating is a float value

# Filter the DataFrame to include only rows that match the user's criteria
filtered_df = df[(df['cuisines type'].str.contains(pr_cuisine, case=False)) & (df['avg cost (two people)'] == pr_rating)]

if not filtered_df.empty:
    # Sort the filtered DataFrame by ratings in descending order
    sorted_df = filtered_df.sort_values(by='rate (out of 5)', ascending=False)

    # Print the top 15 restaurants with the highest ratings
    top_restaurants = sorted_df.head(15)
    for index, row in top_restaurants.iterrows():
        print(f"{row['restaurant name']} - Rating: {row['rate (out of 5)']}")
else:
    print("No restaurants found matching the criteria.")
```

```python
import sklearn as sk
from sklearn.ensemble import RandomForestRegressor
from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_absolute_error

selected_cols = ['avg cost (two people)', 'online_order', 'table booking', 'cuisines type', 'area', 'rate (out of 5)']
df_selected = df[selected_cols]
x = df_selected[['avg cost (two people)', 'online_order', 'table booking', 'cuisines type', 'area' ]]
y =  df.iloc[0:, 4:5].values
column_trans = ColumnTransformer(
    [('encoder', OneHotEncoder(), ['online_order', 'table booking', 'cuisines type', 'area' ])],
    remainder='passthrough')

x_encoded = column_trans.fit_transform(x)
print(x.shape, y.shape)
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y, test_size=0.33, random_state=1)
y_train = y_train.ravel()
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
sk.ensemble.RandomForestRegressor
model = RandomForestRegressor(random_state=1)
model.fit(x_train, y_train)
yhat = model.predict(x_test)
print(yhat)
mae = mean_absolute_error(y_test, yhat)
print('MAE: %.3f' % mae)
```

```python
# Get feature importances from the trained model
importances = model.feature_importances_

# Get the names of the features
feature_names = column_trans.get_feature_names_out()

# Create a DataFrame to store feature importances
feature_importances_df = pd.DataFrame(data={'Feature': feature_names, 'Importance': importances})

# Sort feature importances in descending order
feature_importances_df = feature_importances_df.sort_values(by='Importance', ascending=False)

# Print the correlation between each feature and ratings
feature_importances_df.head(20)
```

```python
lowcost = []
lc_rating = []
highcost =[]
hc_rating = []
for cost, rating in zip(df['avg cost (two people)'], df['rate (out of 5)']):
    if cost <= 800:
        lowcost.append(cost)
        lc_rating.append(rating)
    else:
        highcost.append(cost)
        hc_rating.append(rating)
print(lowcost)
print(lc_rating)
print(highcost)
print(hc_rating)
```

```python
mean_x1 = np.mean(lc_rating)
mean_x2 = np.mean(hc_rating)
print(f"Mean 1: {mean_x1}")
print(f"Mean 2: {mean_x2}")
```

```python
std_x1 = np.std(lc_rating)
std_x2 = np.std(hc_rating)
print(f"Standard deviation 1: {std_x1}")
print(f"Standard deviation 2: {std_x2}")
```

```python
import math
deno_sq = (((std_x1)**2) / len(lc_rating)) + (((std_x2)**2) / len(hc_rating))
deno = math.sqrt(deno_sq)
print(deno)
```

```python
#Test Statistic
t = (mean_x1 - mean_x2)/(deno)
print(t)
```
