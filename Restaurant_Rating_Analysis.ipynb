# **Recommendation System:**
## On the basis of cuisine type and average cost for two

```python
pr_cuisine = input("Enter your preference of cuisine: ")
pr_rating = float(input("Enter the average cost rating for two: "))  # Assuming the rating is a float value

# Filter the DataFrame to include only rows that match the user's criteria
filtered_df = df[(df['cuisines type'].str.contains(pr_cuisine, case=False)) & (df['avg cost (two people)'] == pr_rating)]

if not filtered_df.empty:
    # Sort the filtered DataFrame by ratings in descending order
    sorted_df = filtered_df.sort_values(by='rate (out of 5)', ascending=False)

    # Print the top 15 restaurants with the highest ratings
    top_restaurants = sorted_df.head(15)
    for index, row in top_restaurants.iterrows():
        print(f"{row['restaurant name']} - Rating: {row['rate (out of 5)']}")
else:
    print("No restaurants found matching the criteria.")
```

# **Machine Learning Model**
Training the model to predict the restaurant ratings based on 5 different columns and finding out the features with the highest impact on ratings

```python
import sklearn as sk
from sklearn.ensemble import RandomForestRegressor
from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_absolute_error

selected_cols = ['avg cost (two people)', 'online_order', 'table booking', 'cuisines type', 'area', 'rate (out of 5)']
df_selected = df[selected_cols]
x = df_selected[['avg cost (two people)', 'online_order', 'table booking', 'cuisines type', 'area' ]]
y =  df.iloc[0:, 4:5].values
column_trans = ColumnTransformer(
    [('encoder', OneHotEncoder(), ['online_order', 'table booking', 'cuisines type', 'area' ])],
    remainder='passthrough')

x_encoded = column_trans.fit_transform(x)
print(x.shape, y.shape)
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y, test_size=0.33, random_state=1)
y_train = y_train.ravel()
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
sk.ensemble.RandomForestRegressor
model = RandomForestRegressor(random_state=1)
model.fit(x_train, y_train)
yhat = model.predict(x_test)
print(yhat)
mae = mean_absolute_error(y_test, yhat)
print('MAE: %.3f' % mae)
```

```python
# Get feature importances from the trained model
importances = model.feature_importances_

# Get the names of the features
feature_names = column_trans.get_feature_names_out()

# Create a DataFrame to store feature importances
feature_importances_df = pd.DataFrame(data={'Feature': feature_names, 'Importance': importances})

# Sort feature importances in descending order
feature_importances_df = feature_importances_df.sort_values(by='Importance', ascending=False)

# Print the correlation between each feature and ratings
feature_importances_df.head(20)
```

# **Hypothesis Testing**
## **Z-Test**

Null Hypothesis: The mean rating of restaurants with average cost (for two) less than or equal to 400 is equal to the mean rating of restaurants with average cost for two above 400.


Alternative Hypothesis: The mean rating of restaurants with average cost for two less than or equal to 400 is greater than the mean rating of restaurants with average cost for two above 400.

Level of Significance: 95% or 0.95

Critical Value: 1.96


```python
def calculate_mean_std(df):
    mean_rate = df['rate (out of 5)'].mean()
    std_rate = df['rate (out of 5)'].std()
    return mean_rate, std_rate
```

```python
def calculate_mean_std_condition(df, condition):
    if condition == 'less_than_400':
        filtered_data = df[df['avg cost (two people)'] <= 400]
    elif condition == 'greater_than_400':
        filtered_data = df[df['avg cost (two people)'] > 400]
    else:
        print("Invalid condition")
        return None

    mean_rate, std_rate = calculate_mean_std(filtered_data)
    return mean_rate, std_rate
```

```python
# Calculate mean and standard deviation for average cost less than or equal to 400
mean_less_than_400, std_less_than_400 = calculate_mean_std_condition(df, 'less_than_400')
print("Mean rate for average cost less than or equal to 400:", mean_less_than_400)
print("Standard deviation of rate for average cost less than or equal to 400:", std_less_than_400)

# Calculate mean and standard deviation for average cost greater than 400
mean_greater_than_400, std_greater_than_400 = calculate_mean_std_condition(df, 'greater_than_400')
print("\nMean rate for average cost greater than 800:", mean_greater_than_400)
print("Standard deviation of rate for average cost greater than 400:", std_greater_than_400)
```

```python
def calculate_mean_std_condition(df, condition):
    if condition == 'less_than_400':
        filtered_data = df[df['avg cost (two people)'] <= 400]
    elif condition == 'greater_than_400':
        filtered_data = df[df['avg cost (two people)'] > 400]
    else:
        print("Invalid condition")
        return None

    mean_rate, std_rate = calculate_mean_std(filtered_data)
    count = filtered_data.shape[0]  # Count of rows
    return mean_rate, std_rate, count
```

```python
# Calculate mean and standard deviation for average cost less than or equal to 400
mean_less_than_400, std_less_than_400, count_less_than_400 = calculate_mean_std_condition(df, 'less_than_400')
print("Mean rate for average cost less than or equal to 400:", mean_less_than_400)
print("Standard deviation of rate for average cost less than or equal to 400:", std_less_than_400)
print("Count of average cost less than or equal to 400:", count_less_than_400)

# Calculate mean and standard deviation for average cost greater than 400
mean_greater_than_400, std_greater_than_400, count_greater_than_400 = calculate_mean_std_condition(df, 'greater_than_400')
print("\nMean rate for average cost greater than 400:", mean_greater_than_400)
print("Standard deviation of rate for average cost greater than 400:", std_greater_than_400)
print("Count of average cost greater than 400:", count_greater_than_400)
```

```python
import math
x=mean_less_than_400-mean_greater_than_400
s=(((std_less_than_400)**2)/count_less_than_400) + ((std_greater_than_400**2)/count_greater_than_400)
y=math.sqrt(s)
```

```python
t=x/y
print("Test Statistic: ", t)
```
